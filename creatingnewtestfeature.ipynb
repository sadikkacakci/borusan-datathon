```python
test_data = test_data.drop("Unnamed: 0",axis = 1)
test_data
```

```python
test_data["Trafikteki Gün Sayısı"] = None
test_data["Referans Tarih"] = ["2021-12-31"]*test_data.shape[0]
test_data["InvoiceDate"] = pd.to_datetime(test_data["InvoiceDate"].values)
test_data["Referans Tarih"] = pd.to_datetime(test_data["Referans Tarih"].values)
test_data
```

```python
test_data["Trafikteki Gün Sayısı"] = test_data["Referans Tarih"] - test_data["InvoiceDate"]
test_data['Trafikteki Gün Sayısı'] = test_data['Trafikteki Gün Sayısı'].astype(str)
test_data['Trafikteki Gün Sayısı'] = test_data['Trafikteki Gün Sayısı'].str.replace('days', '')
test_data = test_data.drop("Referans Tarih", axis = 1)
test_data['Trafikteki Gün Sayısı'] = test_data['Trafikteki Gün Sayısı'].astype(int)
test_data
```

```python
test_data = test_data[["City","CustomerId","VehicleId","InvoiceDate","BranchCode","Milage","MainGroupDescription","SubGroupDescription","SplitCode","Split_Grubu","ServiceInvoiceLineNo","ItemDescription","ItemNo","ItemType","MainGroupCode","SubGroupCode","Trafikteki Gün Sayısı","IsMalfunction"]]
test_data
```

```python
test_data["CustomerId"] = test_data["CustomerId"].replace("DAHİLİ","1")
test_data["CustomerId"] = test_data["CustomerId"].replace(np.nan,"0")
test_data["CustomerId"] = test_data["CustomerId"].astype(int)
test_data
```
